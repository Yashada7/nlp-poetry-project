```python
import os
from openai import OpenAI # Using the modern openai library

# --- Configuration ---

client = OpenAI(api_key=os.environ.get("OPENAI_KEY"))

def evaluate_poetry(poem: str, metrics: list) -> str:
    """
    Sends a poem to an AI API for evaluation based on specified metrics.

    Args:
        poem: The text of the poem to be evaluated.
        metrics: A list of metrics (e.g., "Rhyme Scheme", "Imagery", "Emotional Resonance").

    Returns:
        A structured JSON string of the evaluation.
    """
    # 1. Define the Metrics for Evaluation
    metric_list = ", ".join(metrics)

    # 2. Craft the System and User Prompts
    # The System role sets the AI's persona and rules (e.g., output format).
    system_prompt = (
        "You are an expert poetry critic. Your task is to evaluate the user's poem "
        f"based on the following metrics: {metric_list}. "
        "Provide a score from 1 to 10 for each metric and a brief explanation. "
        "Your final output MUST be a JSON object with 'title', 'overall_score', and 'metric_scores' keys."
    )

    # The User role provides the input data (the poem) and final instructions.
    user_prompt = (
        "Please evaluate the following poem:\n\n"
        f"--- POEM START ---\n{poem}\n--- POEM END ---\n\n"
        "Provide your evaluation in the requested JSON format."
    )

    # 3. Call the API
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini", # A powerful and cost-effective model for this task
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt}
            ],
            response_format={"type": "json_object"}, # Enforce JSON output
            temperature=0.3 # Lower temperature for more deterministic, critical output
        )

        return response.choices[0].message.content

    except Exception as e:
        return f"An error occurred: {e}"

# --- Usage Example ---
my_poem = """
In a small town between mountains where mountains lie," â€”Gone to the mountains, where the towns and cities lie, So I was brought there by a young friend whom I never knew." â€”No one knew that other, so I was brought there by a young friend who had never spoken to me. For I had never seen him, I knew that no one could see him, And I have come because he has come to visit me; The land has become so lovely that I have come to see it."â€”for Robert Frost, 1949 The Poem I. Coyote and Nellie went to the desert and had fun together. It was a very pleasant visit, but I do remember some moments of it. I remember

"""

# Define your desired evaluation metrics
evaluation_metrics = [
    "Imagery and Figurative Language",
    "Emotional Resonance",
    "Adherence to Form (e.g., Haiku, Sonnet, Free Verse)",
    "Pacing and Flow"
]

evaluation_result = evaluate_poetry(my_poem, evaluation_metrics)

print("### Evaluation Result (JSON Output) ###")
print(evaluation_result)
```
